```python
# ============================================================
# BLOCK 3: High-Intent Direct Material Extraction Engine
# ============================================================

import re
from pypdf import PdfReader

def battery_expert_scan(pdf_path):
    if hasattr(pdf_path, 'read') or (isinstance(pdf_path, str) and (pdf_path.endswith('.pdf') or '/' in pdf_path or '\\' in pdf_path)):
        reader = PdfReader(pdf_path)
        text = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
        display_name = pdf_path.replace("_MOESM1_ESM.pdf","").replace("_"," ").split('/')[-1].split('\\')[-1]
    else:
        text = str(pdf_path)
        display_name = "Extracted Literature Context"

    # Step 1: Scan for absolute keyword matches across the vocabulary
    all_known_materials = list(MATERIAL_LIBRARY.keys())
    detected_materials = [mat for mat in all_known_materials if re.search(rf'\b{re.escape(mat)}\b', text, re.IGNORECASE)]

    # Explicit role association keywords
    anode_kws = ["anode", "negative electrode", "counter electrode", "lithium foil"]
    cathode_kws = ["cathode", "positive electrode", "working electrode", "composite cathode"]
    se_kws = ["electrolyte", "solid electrolyte", "separator", "se layer"]

    def score_material_intent(material_name, role_keywords):
        """
        Calculates raw material priority based on overall usage count combined
        with strict immediate sentence proximity to intentional keywords.
        """
        # Baseline: Total count of this material in the entire paper
        total_mentions = len(re.findall(rf'\b{re.escape(material_name)}\b', text, re.IGNORECASE))
        if total_mentions == 0:
            return 0

        # Proximity score: Look for explicit role context within a 120-character frame
        proximity_matches = 0
        for kw in role_keywords:
            pattern = rf'(?:{kw})[^.]{{0,120}}{re.escape(material_name)}|{re.escape(material_name)}[^.]{{0,120}}(?:{kw})'
            proximity_matches += len(re.findall(pattern, text, re.IGNORECASE))

        # Weight proximity heavily over passive text mentions
        return (proximity_matches * 15) + total_mentions

    # Map scores dynamically
    anode_scores = {}
    cathode_scores = {}
    se_scores = {}

    for mat in detected_materials:
        m_type = MATERIAL_LIBRARY[mat]["type"]
        if m_type == "anode":
            anode_scores[mat] = score_material_intent(mat, anode_kws)
        elif m_type == "cathode":
            cathode_scores[mat] = score_material_intent(mat, cathode_kws)
        elif m_type == "SE":
            se_scores[mat] = score_material_intent(mat, se_kws)

    # Resolve primary active stack selections (highest overall intent score wins)
    primary_anode = max(anode_scores, key=anode_scores.get) if anode_scores else "N/A"
    primary_cathode = max(cathode_scores, key=cathode_scores.get) if cathode_scores else "N/A"
    primary_se = max(se_scores, key=se_scores.get) if se_scores else "N/A"

    primary_stack = [primary_anode, primary_cathode, primary_se]

    # Map materials directly to active roster
    materials_used = []
    if primary_anode != "N/A":
        meta = MATERIAL_LIBRARY[primary_anode]
        materials_used.append({"name": primary_anode, "role": "Primary Anode", "modulus": meta.get("modulus","—"), "cond": meta.get("sigma_e","—"), "thcap": meta.get("th_cap"), "ewin": None})
    if primary_se != "N/A":
        meta = MATERIAL_LIBRARY[primary_se]
        materials_used.append({"name": primary_se, "role": "Solid Electrolyte Matrix", "modulus": meta.get("modulus","—"), "cond": meta.get("sigma_ion","—"), "thcap": None, "ewin": meta.get("e_window")})
    if primary_cathode != "N/A":
        meta = MATERIAL_LIBRARY[primary_cathode]
        materials_used.append({"name": primary_cathode, "role": "Primary Cathode", "modulus": meta.get("modulus","—"), "cond": meta.get("sigma_ion","—"), "thcap": meta.get("th_cap"), "ewin": None})

    # Compile the Supplementary Context arrays (everything found but NOT running inside the core battery cell)
    supplementary_materials = []
    for mat in detected_materials:
        if mat not in primary_stack:
            meta = MATERIAL_LIBRARY[mat]
            supplementary_materials.append({
                "name": mat,
                "role": f"Supplementary {meta['type'].upper()}"
            })

    # Pull system configurations
    system_class = identify_system_context([MATERIAL_LIBRARY[m] for m in primary_stack if m != "N/A"])

    # Simple, direct parsing for raw mass composition weight ratios
    frac_hits = re.findall(r'(\d+\.?\d*)\s*(?:wt\.?\s*%|vol\.?\s*%|%)', text)
    valid_fracs = [float(v) for v in frac_hits if 0 < float(v) <= 100]

    config = {
        "anode": [{"label": primary_anode if primary_anode != "N/A" else "Anode", "pct": 100.0}],
        "se": [{"label": primary_se if primary_se != "N/A" else "SE Layer", "pct": 100.0}],
        "cathode": [{"label": primary_cathode if primary_cathode != "N/A" else "Cathode Layer", "pct": 100.0}]
    }

    # If an exact composite blend is discovered in the text logs (e.g., 70wt%, 25wt%, 5wt%)
    if len(valid_fracs) >= 3:
        config["cathode"] = [
            {"label": primary_cathode if primary_cathode != "N/A" else "Active Cathode", "pct": valid_fracs[0]},
            {"label": primary_se if primary_se != "N/A" else "SE Matrix", "pct": valid_fracs[1]},
            {"label": "Carbon/Binder Matrix", "pct": valid_fracs[2]}
        ]

    # Quick metric extractors
    def parse_first_match(pattern, fallback="N/A"):
        m = re.findall(pattern, text, re.IGNORECASE)
        return list(set(m))[0] if m else fallback

    return {
        "paper": display_name,
        "sub": " · ".join([m for m in primary_stack if m != "N/A"]),
        "materials": materials_used,
        "supplementary": supplementary_materials,
        "config": config,
        "temp": parse_first_match(r'(\d+\s*°C)'),
        "pressure": parse_first_match(r'(\d+\.?\d*\s*MPa)'),
        "voltage": parse_first_match(r'(\d+\.?\d*\s*[-–]\s*\d+\.?\d*\s*V)'),
        "areal_mah": f"{parse_first_match(r'(\d+\.?\d*)\s*mAh\s*/\s*cm\s*[²2]')} mAh/cm²" if parse_first_match(r'(\d+\.?\d*)\s*mAh\s*/\s*cm\s*[²2]') != "N/A" else "N/A",
        "areal_mg": f"{parse_first_match(r'(\d+\.?\d*)\s*mg\s*/\s*cm\s*[²2]')} mg/cm²" if parse_first_match(r'(\d+\.?\d*)\s*mg\s*/\s*cm\s*[²2]') != "N/A" else "N/A",
        "crates": ", ".join(list(set(re.findall(r'(\d+\.?\d*\s*C)(?!\w)', text))))[:3] or "N/A",
        "retention": parse_first_match(r'(\d+\.?\d*)\s*%'),
        "cycles": parse_first_match(r'(\d+)\s*cycles?'),
        "energy": parse_first_match(r'(\d+\.?\d*\s*Wh\s*/?\s*(?:kg|L))'),
        "system_class": system_class
    }
```

```python
# ============================================================
# BLOCK 3A: Strict Composition Mass Balancer Engine
# ============================================================

def predict_missing_components(existing_segments, layer_type, active_name="Active Material", se_name="Solid Electrolyte"):
    segs = [dict(s) for s in existing_segments]
    current_sum = sum(s["pct"] for s in segs)

    # If values sum cleanly to 100%, change nothing and return
    if abs(current_sum - 100.0) < 1.0 or current_sum <= 0:
        return segs

    remainder = 100.0 - current_sum
    labels_lower = [str(s["label"]).lower() for s in segs]

    # Component fingerprint flags
    has_active = any(any(x in l for x in ["nmc", "lfp", "lco", "nca", "li", "silicon", "graphite", "active"]) for l in labels_lower)
    has_se     = any(any(x in l for x in ["sse", "electrolyte", "solid", "matrix"]) for l in labels_lower)
    has_carbon = any(any(x in l for x in ["carbon", "black", "cnt", "conductive"]) for l in labels_lower)

    # If a composite composition matrix has a minor remaining piece missing, calculate it directly
    if len(segs) > 1:
        if not has_carbon and remainder <= 15.0:
            segs.append({"label": "Carbon C-Additive (Calculated)", "pct": remainder, "style": "color:#10b981; font-style:italic;"})
        elif not has_se and se_name and remainder <= 40.0:
            segs.append({"label": f"{se_name} Matrix (Calculated)", "pct": remainder, "style": "color:#0284c7;"})
        else:
            segs.append({"label": "Auxiliary Matrix (Predicted)", "pct": remainder, "style": "color:#6b7280;"})
        return segs

    # Default fallback handling for single unassigned layers
    if "cathode" in layer_type.lower():
        return [
            {"label": active_name if active_name else "Active Cathode", "pct": 70.0},
            {"label": se_name if se_name else "Solid Electrolyte", "pct": 25.0},
            {"label": "Carbon Black / Binder", "pct": 5.0}
        ]
    elif "anode" in layer_type.lower():
        return [
            {"label": active_name if active_name else "Active Anode", "pct": 95.0},
            {"label": "Conductive Carbon Binder", "pct": 5.0}
        ]

    return [{"label": se_name if se_name else "Solid Electrolyte", "pct": 100.0}]
```

```python
# ============================================================
# BLOCK 3: High-Intent Direct Material Extraction Engine
# ============================================================

import re
from pypdf import PdfReader

def battery_expert_scan(pdf_path):
    # Safe handling for both structural PDF paths and raw context text inputs
    if hasattr(pdf_path, 'read') or (isinstance(pdf_path, str) and (pdf_path.endswith('.pdf') or '/' in pdf_path or '\\' in pdf_path)):
        reader = PdfReader(pdf_path)
        text = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
        display_name = pdf_path.replace("_MOESM1_ESM.pdf","").replace("_"," ").split('/')[-1].split('\\')[-1]
    else:
        text = str(pdf_path)
        display_name = "Extracted Literature Context"

    # Ingest baseline vocabulary mapping list
    all_known_materials = list(MATERIAL_LIBRARY.keys())
    detected_materials = [mat for mat in all_known_materials if re.search(rf'\b{re.escape(mat)}\b', text, re.IGNORECASE)]

    # Explicit role association parameters
    anode_kws = ["anode", "negative electrode", "counter electrode", "lithium foil"]
    cathode_kws = ["cathode", "positive electrode", "working electrode", "composite cathode"]
    se_kws = ["electrolyte", "solid electrolyte", "separator", "se layer"]

    def score_material_intent(material_name, role_keywords):
        """
        Calculates focus weight parameters based on cumulative frequency metrics
        combined with strict look-ahead syntax proximity mapping frames.
        """
        total_mentions = len(re.findall(rf'\b{re.escape(material_name)}\b', text, re.IGNORECASE))
        if total_mentions == 0:
            return 0

        proximity_matches = 0
        for kw in role_keywords:
            pattern = rf'(?:{kw})[^.]{{0,120}}{re.escape(material_name)}|{re.escape(material_name)}[^.]{{0,120}}(?:{kw})'
            proximity_matches += len(re.findall(pattern, text, re.IGNORECASE))

        return (proximity_matches * 15) + total_mentions

    # Isolate and score candidates dynamically by categorical types
    anode_scores = {mat: score_material_intent(mat, anode_kws) for mat in detected_materials if MATERIAL_LIBRARY[mat]["type"] == "anode"}
    cathode_scores = {mat: score_material_intent(mat, cathode_kws) for mat in detected_materials if MATERIAL_LIBRARY[mat]["type"] == "cathode"}
    se_scores = {mat: score_material_intent(mat, se_kws) for mat in detected_materials if MATERIAL_LIBRARY[mat]["type"] == "SE"}

    # Resolve primary active stack selections (highest structural intent score wins)
    primary_anode = max(anode_scores, key=anode_scores.get) if anode_scores else "N/A"
    primary_cathode = max(cathode_scores, key=cathode_scores.get) if cathode_scores else "N/A"
    primary_se = max(se_scores, key=se_scores.get) if se_scores else "N/A"

    primary_stack = [primary_anode, primary_cathode, primary_se]

    # Map core data dictionary to active rows
    materials_used = []
    if primary_anode != "N/A":
        meta = MATERIAL_LIBRARY[primary_anode]
        materials_used.append({"name": primary_anode, "role": "Anode active", "modulus": meta.get("modulus","—"), "cond": meta.get("sigma_e","—"), "thcap": meta.get("th_cap"), "ewin": None})
    if primary_se != "N/A":
        meta = MATERIAL_LIBRARY[primary_se]
        materials_used.append({"name": primary_se, "role": "Solid electrolyte", "modulus": meta.get("modulus","—"), "cond": meta.get("sigma_ion","—"), "thcap": None, "ewin": meta.get("e_window")})
    if primary_cathode != "N/A":
        meta = MATERIAL_LIBRARY[primary_cathode]
        materials_used.append({"name": primary_cathode, "role": "Cathode active", "modulus": meta.get("modulus","—"), "cond": meta.get("sigma_ion","—"), "thcap": meta.get("th_cap"), "ewin": None})

    # Route everything found elsewhere in text straight into the comprehensive Supplementary roster
    supplementary_materials = [{"name": m, "role": MATERIAL_LIBRARY[m]["type"]} for m in detected_materials if m not in primary_stack]

    # ── STAGE II: Targeted Cathode Ratio Extraction ──
    # Isolates a contextual search frame around cathode descriptions to eliminate background metric noise
    cathode_context = ""
    cathode_matches = list(re.finditer(r'cathode|positive electrode|weight|ratio', text, re.IGNORECASE))
    for m in cathode_matches:
        start = max(0, m.start() - 150)
        end = min(len(text), m.end() + 250)
        cathode_context += text[start:end] + " | "

    # Sweep for layout percentage values inside the target cathode contextual scope
    raw_ratios = [float(x) for x in re.findall(r'(\d+\.?\d*)\s*(?:wt\.?\s*%|vol\.?\s*%|%)', cathode_context)]

    # Filter out outlier values highly unlikely to track composition weights (e.g., temperatures/voltages)
    valid_mass_fractions = [v for v in raw_ratios if 5.0 <= v <= 90.0][:3]

    # Fallback template defaults if composition splits are unresolvable from character arrays
    if len(valid_mass_fractions) < 2:
        valid_mass_fractions = [70.0, 25.0, 5.0]

    config = {
        "anode": [{"label": primary_anode if primary_anode != "N/A" else "Anode Active", "pct": 100.0}],
        "se": [{"label": primary_se if primary_se != "N/A" else "SE Layer", "pct": 100.0}],
        "cathode": []
    }

    # Linearly bundle ratios to structural component tags
    labels = [
        primary_cathode if primary_cathode != "N/A" else "Active Material",
        primary_se if primary_se != "N/A" else "Solid Electrolyte",
        "Carbon Black / Binder Additives"
    ]

    for idx, val in enumerate(valid_mass_fractions):
        if idx < len(labels):
            config["cathode"].append({"label": labels[idx], "pct": val})

    # Helper utility for single-hit metric parsing sweeps
    def parse_first_match(pattern, fallback="N/A"):
        m = re.findall(pattern, text, re.IGNORECASE)
        return list(set(m))[0] if m else fallback

    return {
        "paper": display_name,
        "sub": " · ".join([m for m in primary_stack if m != "N/A"]),
        "materials": materials_used,
        "supplementary": supplementary_materials,
        "config": config,
        "temp": parse_first_match(r'(\d+\s*°C)'),
        "pressure": parse_first_match(r'(\d+\.?\d*\s*MPa)'),
        "voltage": parse_first_match(r'(\d+\.?\d*\s*[-–]\s*\d+\.?\d*\s*V)'),
        "areal_mah": f"{parse_first_match(r'(\d+\.?\d*)\s*mAh\s*/\s*cm\s*[²2]')} mAh/cm²" if parse_first_match(r'(\d+\.?\d*)\s*mAh\s*/\s*cm\s*[²2]') != "N/A" else "N/A",
        "areal_mg": f"{parse_first_match(r'(\d+\.?\d*)\s*mg\s*/\s*cm\s*[²2]')} mg/cm²" if parse_first_match(r'(\d+\.?\d*)\s*mg\s*/\s*cm\s*[²2]') != "N/A" else "N/A",
        "crates": ", ".join(list(set(re.findall(r'(\d+\.?\d*\s*C)(?!\w)', text))))[:3] or "N/A",
        "retention": parse_first_match(r'(\d+\.?\d*)\s*%'),
        "cycles": parse_first_match(r'(\d+)\s*cycles?'),
        "energy": parse_first_match(r'(\d+\.?\d*\s*Wh\s*/?\s*(?:kg|L))'),
    }
```

```python
# ============================================================
# BLOCK 4: Unified UI Production Render Engine
# ============================================================

from IPython.display import display, HTML
import json
import math

SECTION_PALETTES = {
    "COMPOSITE ANODE":     ["#1A5276", "#2980B9", "#5DADE2", "#AED6F1"],
    "SEPARATOR / SE":      ["#117864", "#16A085", "#48C9B0", "#A3E4D7"],
    "COMPOSITE CATHODE":   ["#A04000", "#D35400", "#E59866", "#F5CBA7"],
    "Default":             ["#4D5656", "#7F8C8D", "#BDC3C7", "#EAEDED"]
}

def donut_svg(segments, size=52, palette=None):
    if palette is None:
        palette = SECTION_PALETTES["Default"]

    cx = cy = size / 2
    r, inner = size * 0.38, size * 0.22

    segs = [dict(s) for s in segments if s.get("pct", 0) > 0]
    total_raw_sum = sum(s["pct"] for s in segs)

    if not segs or total_raw_sum <= 0:
        return f'<svg width="{size}" height="{size}"><circle cx="{cx}" cy="{cy}" r="{r}" fill="none" stroke="#e0e0e0" stroke-width="{r-inner}"/></svg>'

    if len(segs) == 1:
        return f'<svg width="{size}" height="{size}" style="transform: rotate(-90deg);"><circle cx="{cx}" cy="{cy}" r="{r}" fill="none" stroke="{palette[0]}" stroke-width="{r-inner}"/></svg>'

    angle = -math.pi / 2
    paths = ""
    for i, seg in enumerate(segs):
        normalized_fraction = seg["pct"] / total_raw_sum
        a = normalized_fraction * 2 * math.pi
        end = angle + a

        x1, y1 = cx + r * math.cos(angle), cy + r * math.sin(angle)
        x2, y2 = cx + r * math.cos(end),   cy + r * math.sin(end)
        ix1, iy1 = cx + inner * math.cos(angle), cy + inner * math.sin(angle)
        ix2, iy2 = cx + inner * math.cos(end),   cy + inner * math.sin(end)
        large = 1 if a > math.pi else 0

        paths += f'<path d="M{ix1:.1f},{iy1:.1f} L{x1:.1f},{y1:.1f} A{r:.1f},{r:.1f} 0 {large} 1 {x2:.1f},{y2:.1f} L{ix2:.1f},{iy2:.1f} A{inner:.1f},{inner:.1f} 0 {large} 0 {ix1:.1f},{iy1:.1f}Z" fill="{palette[i % len(palette)]}"/>'
        angle = end
    return f'<svg width="{size}" height="{size}">{paths}</svg>'

def config_html(cfg, primary_cathode, primary_anode, primary_se, supplementary_list):
    sections = [
        ("COMPOSITE ANODE",   cfg.get("anode",[]),   primary_anode,   primary_se),
        ("SEPARATOR / SE",    cfg.get("se",[]),      primary_se,      None),
        ("COMPOSITE CATHODE", cfg.get("cathode",[]), primary_cathode, primary_se),
    ]
    html = '<div style="display:flex; flex-direction:column; gap:10px;">'
    for label, segs, act_mat, se_mat in sections:
        if not segs: continue

        local_segs = [dict(s) for s in segs]
        total_sum = sum(s["pct"] for s in local_segs)

        if "anode" in label.lower() and act_mat in ["Li metal", "lithium foil"]:
            local_segs = [{"label": f"{act_mat}", "pct": 100.0}]
            total_sum = 100.0

        html += f'<div><div style="font-size:10px; font-weight:600; color:#444; text-transform:uppercase; letter-spacing:0.5px; margin-bottom:4px;">{label}</div>'
        html += '<div style="display:flex; align-items:center; gap:14px;">'
        palette = SECTION_PALETTES.get(label, SECTION_PALETTES["Default"])
        html += donut_svg(local_segs, palette=palette)
        html += '<div style="flex:1">'

        for j, s in enumerate(local_segs):
            col = palette[j % len(palette)]
            display_pct = (s["pct"] / total_sum) * 100 if total_sum > 0 else 0

            html += (f'<div style="display:flex; align-items:center; justify-content:space-between; gap:6px; margin:2px 0;">'
                     f'<div style="display:flex; align-items:center; gap:6px; overflow:hidden;">'
                     f'<div style="width:8px; height:8px; border-radius:2px; background:{col}; flex-shrink:0;"></div>'
                     f'<span style="font-size:10px; color:#444; white-space:nowrap; text-overflow:ellipsis;">{s["label"]}</span></div>'
                     f'<span style="font-size:10px; font-weight:600; color:#111;">{display_pct:.0f}%</span></div>')
        html += '</div></div></div>'
        if label != "COMPOSITE CATHODE":
            html += '<div style="border-top:0.5px solid #e0e0e0; margin:4px 0;"></div>'

    # Dynamic Multi-line Grid block for supplementary elements directly below the charts
    html += '<div style="border-top:1px solid #ccc; margin-top:8px; padding-top:6px;">'
    html += '<div style="font-size:9px; font-weight:700; color:#555; text-transform:uppercase; letter-spacing:0.5px; margin-bottom:6px;">Supplementary Materials Mentioned</div>'
    if supplementary_list:
        badges = "".join([f'<span style="background:#f1f5f9; color:#475569; font-size:9px; font-weight:600; padding:3px 6px; border-radius:3px; border:1px solid #cbd5e1; display:inline-block; margin-bottom:4px; margin-right:2px;">{s["name"]}</span>' for s in supplementary_list])
        html += f'<div style="display:block; width:100%; line-height:1.6;">{badges}</div>'
    else:
        html += '<div style="font-size:10px; color:#aaa; font-style:italic;">None identified in text</div>'
    html += '</div>'

    return html + '</div>'

def kv(k, v):
    if not v or v in ["N/A", "—"] or v.startswith("N/A"): return ""
    return f'<div style="display:flex; justify-content:space-between; align-items:center; gap:12px; margin:3px 0;"><span style="font-size:10px; font-weight:500; color:#777;">{k}</span><span style="font-size:10px; font-weight:600; color:#222; text-align:right;">{v}</span></div>'

def ret_color(r):
    try:
        v = float(r.replace("%",""))
        return "#085041" if v>=90 else "#633806" if v>=80 else "#712B13"
    except: return "inherit"

def material_sort_key(m):
    role = str(m.get("role", "")).lower()
    if "anode" in role: return 1
    if "se" in role or "separator" in role or "electrolyte" in role: return 2
    if "cathode" in role: return 3
    return 4

def build_rows(results):
    html = ""
    for pi, p in enumerate(results):
        mats = p["materials"] or []
        mats = sorted(mats, key=material_sort_key)
        span = len(mats)

        c_name = next((mat["name"] for mat in mats if "cathode" in str(mat.get("role","")).lower()), "Cathode")
        a_name = next((mat["name"] for mat in mats if "anode" in str(mat.get("role","")).lower()), "Anode")
        se_name = next((mat["name"] for mat in mats if "electrolyte" in str(mat.get("role","")).lower()), "SSE")

        for mi, m in enumerate(mats):
            is_first = (mi == 0)
            top_border = "2px solid #ccc" if pi > 0 else "0.5px solid #e0e0e0"
            row_top = top_border if is_first else "0.5px solid #e0e0e0"
            cell_style = f"border-top:{row_top}; padding:10px 12px; vertical-align:top; border-right:0.5px solid #e0e0e0; border-bottom:0.5px solid #e0e0e0;"
            shared_style = f"border-top:{top_border}; padding:10px 12px; vertical-align:top; border-right:0.5px solid #e0e0e0; border-bottom:0.5px solid #e0e0e0;"

            html += "<tr>"
            if is_first:
                html += f'<td rowspan="{span}" style="{shared_style}"><div style="font-size:11px; font-weight:700; color:#111;">{p["paper"]}</div><div style="font-size:10px; color:#888; margin-top:4px; line-height:1.4;">{p["sub"]}</div></td>'

            html += f'<td style="{cell_style}"><div style="font-size:11px; font-weight:700; color:#111;">{m["name"]}</div><div style="font-size:10px; color:#aaa; margin-top:2px;">{m["role"]}</div></td>'
            prop_html = kv("Modulus", m["modulus"]) + kv("Conductivity", m["cond"]) + kv("Th. capacity", m.get("thcap")) + kv("E window", m.get("ewin"))
            html += f'<td style="{cell_style}">{prop_html}</td>'

            if is_first:
                rc = ret_color(p["retention"])
                ret_cell_style = f"border-top:{top_border}; padding:10px 12px; vertical-align:middle; text-align:center; border-right:0.5px solid #e0e0e0; border-bottom:0.5px solid #e0e0e0;"

                html += f'<td rowspan="{span}" style="{shared_style}">{config_html(p["config"], c_name, a_name, se_name, p["supplementary"])}</td>'
                testing_html = f'<div style="margin-bottom:2px;">{kv("Temp", p["temp"])}{kv("Pressure", p["pressure"])}<div style="border-top:0.5px dashed #e0e0e0; margin:6px 0;"></div>{kv("Mass Areal", p["areal_mg"])}{kv("Cap Areal", p["areal_mah"])}<div style="border-top:0.5px dashed #e0e0e0; margin:6px 0;"></div>{kv("C-Rate", p["crates"])}{kv("Voltage", p["voltage"])}</div>'
                html += f'<td rowspan="{span}" style="{shared_style}">{testing_html}</td>'
                retention_html = f'<div style="font-size:20px; font-weight:700; color:{rc}; text-align:center;">{p["retention"]}</div><div style="font-size:10px; color:#888; margin-top:3px; text-align:center;">@ {p["cycles"]} cycles</div>'
                html += f'<td rowspan="{span}" style="{ret_cell_style}">{retention_html}</td>'
                energy_html = f'{kv("Total Cycles", p["cycles"] + " cyc")}<div style="border-top:0.5px dashed #e0e0e0; margin:5px 0;"></div>{kv("Spec. Energy", p.get("energy"))}<div style="font-size:9px; color:#999; text-align:center; font-style:italic; margin-top:6px; border-top:0.5px dashed #e0e0e0; padding-top:4px;">Basis: Cell-level</div>'
                html += f'<td rowspan="{span}" style="{shared_style}">{energy_html}</td>'

            html += "</tr>"
    return html

TABLE_CSS = """
<style>
  .btbl * { box-sizing:border-box; margin:0; padding:0; font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif; }
  .btbl { overflow-x:auto; padding:1rem; background:#fff; }
  .btbl table { border-collapse:collapse; width:100%; min-width:1150px; border:0.5px solid #e0e0e0; }
  .btbl th { background:#f7f6f3; font-size:11px; font-weight:500; color:#555; padding:9px 12px; border:0.5px solid #e0e0e0; text-align:left; }
  .btbl .gh { text-align:center; font-size:12px; font-weight:600; color:#1a1a18; background:#f4f3ef; }
  .btbl td { font-size:11px; color:#1a1a18; }
</style>
"""

def build_table(results):
    return TABLE_CSS + f"""
<div class="btbl">
<table>
  <thead>
    <tr>
      <th rowspan="2" style="width:14%; vertical-align:bottom;">Paper</th>
      <th colspan="2" class="gh">Materials Info</th>
      <th colspan="2" class="gh">System Level Architecture</th>
      <th colspan="2" class="gh">Performance Indices</th>
    </tr>
    <tr>
      <th style="width:13%">Material</th>
      <th style="width:16%">Modulus &amp; Conductivity</th>
      <th style="width:23%">Battery Configuration</th>
      <th style="width:14%">Testing Conditions</th>
      <th style="width:10%">Retention</th>
      <th style="width:10%">Cycles &amp; Energy</th>
    </tr>
  </thead>
  <tbody>{build_rows(results)}</tbody>
</table>
</div>"""

# --- Execution Entry Point ---
final_results = []
for f in my_files:
    try:
        print(f"Scanning {f} ...")
        final_results.append(battery_expert_scan(f))
        print("  ✓ Done")
    except Exception as e:
        print(f"  ✗ Error: {e}")

display(HTML(build_table(final_results)))
with open("battery_table.html", "w", encoding="utf-8") as f:
    f.write(build_table(final_results))
print("Saved → battery_table.html")
```

```python
# ============================================================
# BLOCK 3: High-Intent Direct Material Extraction Engine
# ============================================================

import re
from pypdf import PdfReader

def battery_expert_scan(pdf_path):
    if hasattr(pdf_path, 'read') or (isinstance(pdf_path, str) and (pdf_path.endswith('.pdf') or '/' in pdf_path or '\\' in pdf_path)):
        reader = PdfReader(pdf_path)
        text = "".join([page.extract_text() for page in reader.pages if page.extract_text()])
        display_name = pdf_path.replace("_MOESM1_ESM.pdf","").replace("_"," ").split('/')[-1].split('\\')[-1]
    else:
        text = str(pdf_path)
        display_name = "Extracted Literature Context"

    all_known_materials = list(MATERIAL_LIBRARY.keys())
    detected_materials = [mat for mat in all_known_materials if re.search(rf'\b{re.escape(mat)}\b', text, re.IGNORECASE)]

    anode_kws = ["anode", "negative electrode", "counter electrode", "lithium foil"]
    cathode_kws = ["cathode", "positive electrode", "working electrode", "composite cathode"]
    se_kws = ["electrolyte", "solid electrolyte", "separator", "se layer"]

    def score_material_intent(material_name, role_keywords):
        total_mentions = len(re.findall(rf'\b{re.escape(material_name)}\b', text, re.IGNORECASE))
        if total_mentions == 0: return 0
        proximity_matches = 0
        for kw in role_keywords:
            pattern = rf'(?:{kw})[^.]{{0,120}}{re.escape(material_name)}|{re.escape(material_name)}[^.]{{0,120}}(?:{kw})'
            proximity_matches += len(re.findall(pattern, text, re.IGNORECASE))
        return (proximity_matches * 15) + total_mentions

    anode_scores = {mat: score_material_intent(mat, anode_kws) for mat in detected_materials if MATERIAL_LIBRARY[mat]["type"] == "anode"}
    cathode_scores = {mat: score_material_intent(mat, cathode_kws) for mat in detected_materials if MATERIAL_LIBRARY[mat]["type"] == "cathode"}
    se_scores = {mat: score_material_intent(mat, se_kws) for mat in detected_materials if MATERIAL_LIBRARY[mat]["type"] == "SE"}

    primary_anode = max(anode_scores, key=anode_scores.get) if anode_scores else "N/A"
    primary_cathode = max(cathode_scores, key=cathode_scores.get) if cathode_scores else "N/A"
    primary_se = max(se_scores, key=se_scores.get) if se_scores else "N/A"

    primary_stack = [primary_anode, primary_cathode, primary_se]

    materials_used = []
    if primary_anode != "N/A":
        meta = MATERIAL_LIBRARY[primary_anode]
        materials_used.append({"name": primary_anode, "role": "Anode active", "modulus": meta.get("modulus","—"), "cond": meta.get("sigma_e","—"), "thcap": meta.get("th_cap"), "ewin": None})
    if primary_se != "N/A":
        meta = MATERIAL_LIBRARY[primary_se]
        materials_used.append({"name": primary_se, "role": "Solid electrolyte", "modulus": meta.get("modulus","—"), "cond": meta.get("sigma_ion","—"), "thcap": None, "ewin": meta.get("e_window")})
    if primary_cathode != "N/A":
        meta = MATERIAL_LIBRARY[primary_cathode]
        materials_used.append({"name": primary_cathode, "role": "Cathode active", "modulus": meta.get("modulus","—"), "cond": meta.get("sigma_ion","—"), "thcap": meta.get("th_cap"), "ewin": None})

    # Separate secondary discoveries cleanly into sorted categories
    supplementary_materials = {"anode": [], "cathode": [], "SE": []}
    for m in detected_materials:
        if m not in primary_stack:
            m_type = MATERIAL_LIBRARY[m]["type"]
            if m_type in supplementary_materials:
                supplementary_materials[m_type].append(m)

    cathode_context = ""
    cathode_matches = list(re.finditer(r'cathode|positive electrode|weight|ratio', text, re.IGNORECASE))
    for m in cathode_matches:
        start = max(0, m.start() - 150)
        end = min(len(text), m.end() + 250)
        cathode_context += text[start:end] + " | "

    raw_ratios = [float(x) for x in re.findall(r'(\d+\.?\d*)\s*(?:wt\.?\s*%|vol\.?\s*%|%)', cathode_context)]
    valid_mass_fractions = [v for v in raw_ratios if 5.0 <= v <= 90.0][:3]

    if len(valid_mass_fractions) < 2:
        valid_mass_fractions = [70.0, 25.0, 5.0]

    config = {
        "anode": [{"label": primary_anode if primary_anode != "N/A" else "Anode Active", "pct": 100.0}],
        "se": [{"label": primary_se if primary_se != "N/A" else "SE Layer", "pct": 100.0}],
        "cathode": []
    }

    labels = [
        primary_cathode if primary_cathode != "N/A" else "Active Material",
        primary_se if primary_se != "N/A" else "Solid Electrolyte",
        "Carbon Black / Binders"
    ]

    for idx, val in enumerate(valid_mass_fractions):
        if idx < len(labels):
            config["cathode"].append({"label": labels[idx], "pct": val})

    def parse_first_match(pattern, fallback="N/A"):
        m = re.findall(pattern, text, re.IGNORECASE)
        return list(set(m))[0] if m else fallback

    return {
        "paper": display_name,
        "sub": " · ".join([m for m in primary_stack if m != "N/A"]),
        "materials": materials_used,
        "supplementary": supplementary_materials,
        "config": config,
        "temp": parse_first_match(r'(\d+\s*°C)'),
        "pressure": parse_first_match(r'(\d+\.?\d*\s*MPa)'),
        "voltage": parse_first_match(r'(\d+\.?\d*\s*[-–]\s*\d+\.?\d*\s*V)'),
        "areal_mah": f"{parse_first_match(r'(\d+\.?\d*)\s*mAh\s*/\s*cm\s*[²2]')} mAh/cm²" if parse_first_match(r'(\d+\.?\d*)\s*mAh\s*/\s*cm\s*[²2]') != "N/A" else "N/A",
        "areal_mg": f"{parse_first_match(r'(\d+\.?\d*)\s*mg\s*/\s*cm\s*[²2]')} mg/cm²" if parse_first_match(r'(\d+\.?\d*)\s*mg\s*/\s*cm\s*[²2]') != "N/A" else "N/A",
        "crates": ", ".join(list(set(re.findall(r'(\d+\.?\d*\s*C)(?!\w)', text))))[:3] or "N/A",
        "retention": parse_first_match(r'(\d+\.?\d*)\s*%'),
        "cycles": parse_first_match(r'(\d+)\s*cycles?'),
        "energy": parse_first_match(r'(\d+\.?\d*\s*Wh\s*/?\s*(?:kg|L))'),
    }
```

```python
# ============================================================
# BLOCK 4: Unified UI Production Render Engine (Complete)
# ============================================================

from IPython.display import display, HTML
import json
import math

# Premium, high-contrast research palettes matching your UI theme
SECTION_PALETTES = {
    "COMPOSITE ANODE":     ["#1A5276", "#2980B9", "#5DADE2", "#AED6F1"],  # Progressive Blues
    "SEPARATOR / SE":      ["#117864", "#16A085", "#48C9B0", "#A3E4D7"],  # Clean Teals
    "COMPOSITE CATHODE":   ["#A04000", "#D35400", "#E59866", "#F5CBA7"],  # High-Voltage Oranges
    "Default":             ["#4D5656", "#7F8C8D", "#BDC3C7", "#EAEDED"]   # Sleek Grays
}

def donut_svg(segments, size=52, palette=None):
    if palette is None:
        palette = SECTION_PALETTES["Default"]
    cx = cy = size / 2
    r, inner = size * 0.38, size * 0.22
    segs = [dict(s) for s in segments if s.get("pct", 0) > 0]

    # Force absolute relative normalization constraints
    total_raw_sum = sum(s["pct"] for s in segs)
    if not segs or total_raw_sum <= 0:
        return f'<svg width="{size}" height="{size}"><circle cx="{cx}" cy="{cy}" r="{r}" fill="none" stroke="#e0e0e0" stroke-width="{r-inner}"/></svg>'
    if len(segs) == 1:
        return f'<svg width="{size}" height="{size}" style="transform: rotate(-90deg);"><circle cx="{cx}" cy="{cy}" r="{r}" fill="none" stroke="{palette[0]}" stroke-width="{r-inner}"/></svg>'

    angle = -math.pi / 2
    paths = ""
    for i, seg in enumerate(segs):
        # NORMALIZATION STEP: Force scale to always map a true mathematical 100% footprint
        a = (seg["pct"] / total_raw_sum) * 2 * math.pi
        end = angle + a
        x1, y1 = cx + r * math.cos(angle), cy + r * math.sin(angle)
        x2, y2 = cx + r * math.cos(end),   cy + r * math.sin(end)
        ix1, iy1 = cx + inner * math.cos(angle), cy + inner * math.sin(angle)
        ix2, iy2 = cx + inner * math.cos(end),   cy + inner * math.sin(end)
        large = 1 if a > math.pi else 0
        paths += f'<path d="M{ix1:.1f},{iy1:.1f} L{x1:.1f},{y1:.1f} A{r:.1f},{r:.1f} 0 {large} 1 {x2:.1f},{y2:.1f} L{ix2:.1f},{iy2:.1f} A{inner:.1f},{inner:.1f} 0 {large} 0 {ix1:.1f},{iy1:.1f}Z" fill="{palette[i % len(palette)]}"/>'
        angle = end
    return f'<svg width="{size}" height="{size}">{paths}</svg>'

def config_html(cfg, primary_cathode, primary_anode, primary_se, supp_dict):
    sections = [
        ("COMPOSITE ANODE",   cfg.get("anode",[]),   primary_anode,   primary_se),
        ("SEPARATOR / SE",    cfg.get("se",[]),      primary_se,      None),
        ("COMPOSITE CATHODE", cfg.get("cathode",[]), primary_cathode, primary_se),
    ]
    html = '<div style="display:flex; flex-direction:column; gap:10px;">'
    for label, segs, act_mat, se_mat in sections:
        if not segs: continue
        local_segs = [dict(s) for s in segs]
        total_sum = sum(s["pct"] for s in local_segs)

        if "anode" in label.lower() and act_mat in ["Li metal", "lithium foil"]:
            local_segs = [{"label": f"{act_mat}", "pct": 100.0}]
            total_sum = 100.0

        html += f'<div><div style="font-size:10px; font-weight:600; color:#444; text-transform:uppercase; letter-spacing:0.5px; margin-bottom:4px;">{label}</div>'
        html += '<div style="display:flex; align-items:center; gap:14px;">'
        palette = SECTION_PALETTES.get(label, SECTION_PALETTES["Default"])
        html += donut_svg(local_segs, palette=palette)
        html += '<div style="flex:1">'
        for j, s in enumerate(local_segs):
            col = palette[j % len(palette)]
            display_pct = (s["pct"] / total_sum) * 100 if total_sum > 0 else 0
            html += (f'<div style="display:flex; align-items:center; justify-content:space-between; gap:6px; margin:2px 0;">'
                     f'<div style="display:flex; align-items:center; gap:6px; overflow:hidden;">'
                     f'<div style="width:8px; height:8px; border-radius:2px; background:{col}; flex-shrink:0;"></div>'
                     f'<span style="font-size:10px; color:#444; white-space:nowrap; text-overflow:ellipsis;">{s["label"]}</span></div>'
                     f'<span style="font-size:10px; font-weight:600; color:#111;">{display_pct:.0f}%</span></div>')
        html += '</div></div></div>'
        if label != "COMPOSITE CATHODE":
            html += '<div style="border-top:0.5px solid #e0e0e0; margin:4px 0;"></div>'

    # Categorized 3-Column Supplementary Matrix Panel directly underneath the charts
    html += '<div style="border-top:1px solid #ccc; margin-top:8px; padding-top:8px;">'
    html += '<div style="font-size:9px; font-weight:700; color:#444; text-transform:uppercase; letter-spacing:0.5px; margin-bottom:6px;">Supplementary References Identified</div>'
    html += '<div style="display:flex; justify-content:space-between; gap:6px; background:#fafafa; padding:6px; border-radius:4px; border:0.5px solid #e2e8f0;">'

    for cat_name, title_lbl in [("anode", "Anodes"), ("cathode", "Cathodes"), ("SE", "Electrolytes")]:
        items_list = supp_dict.get(cat_name, []) if isinstance(supp_dict, dict) else []
        badges_html = "".join([f'<div style="font-size:9px; font-weight:600; color:#475569; background:#fff; border:0.5px solid #cbd5e1; padding:1px 4px; border-radius:2px; margin:2px 0; text-align:center;">{i}</div>' for i in items_list]) or '<div style="font-size:9px; color:#bbb; font-style:italic;">None</div>'
        html += f'<div style="flex:1; display:flex; flex-direction:column;"><div style="font-size:9px; font-weight:700; color:#777; border-bottom:0.5px solid #e2e8f0; padding-bottom:2px; margin-bottom:2px; text-align:center;">{title_lbl}</div>{badges_html}</div>'

    html += '</div></div>'
    return html + '</div>'

def kv(k, v):
    if not v or v in ["N/A", "—"] or v.startswith("N/A"): return ""
    return f'<div style="display:flex; justify-content:space-between; align-items:center; gap:12px; margin:3px 0;"><span style="font-size:10px; font-weight:500; color:#777;">{k}</span><span style="font-size:10px; font-weight:600; color:#222; text-align:right;">{v}</span></div>'

def ret_color(r):
    try:
        v = float(r.replace("%",""))
        return "#085041" if v>=90 else "#633806" if v>=80 else "#712B13"
    except: return "inherit"

def material_sort_key(m):
    role = str(m.get("role", "")).lower()
    if "anode" in role: return 1
    if "se" in role or "separator" in role or "electrolyte" in role: return 2
    if "cathode" in role: return 3
    return 4

def build_rows(results):
    html = ""
    for pi, p in enumerate(results):
        mats = p["materials"] or []
        mats = sorted(mats, key=material_sort_key)
        span = len(mats)

        c_name = next((mat["name"] for mat in mats if "cathode" in str(mat.get("role","")).lower()), "Cathode")
        a_name = next((mat["name"] for mat in mats if "anode" in str(mat.get("role","")).lower()), "Anode")
        se_name = next((mat["name"] for mat in mats if "electrolyte" in str(mat.get("role","")).lower()), "SSE")

        for mi, m in enumerate(mats):
            is_first = (mi == 0)
            top_border = "2px solid #ccc" if pi > 0 else "0.5px solid #e0e0e0"
            row_top = top_border if is_first else "0.5px solid #e0e0e0"
            cell_style = f"border-top:{row_top}; padding:10px 12px; vertical-align:top; border-right:0.5px solid #e0e0e0; border-bottom:0.5px solid #e0e0e0;"
            shared_style = f"border-top:{top_border}; padding:10px 12px; vertical-align:top; border-right:0.5px solid #e0e0e0; border-bottom:0.5px solid #e0e0e0;"

            html += "<tr>"
            if is_first:
                html += f'<td rowspan="{span}" style="{shared_style}"><div style="font-size:11px; font-weight:700; color:#111;">{p["paper"]}</div><div style="font-size:10px; color:#888; margin-top:4px; line-height:1.4;">{p["sub"]}</div></td>'

            html += f'<td style="{cell_style}"><div style="font-size:11px; font-weight:700; color:#111;">{m["name"]}</div><div style="font-size:10px; color:#aaa; margin-top:2px;">{m["role"]}</div></td>'
            prop_html = kv("Modulus", m["modulus"]) + kv("Conductivity", m["cond"]) + kv("Th. capacity", m.get("thcap")) + kv("E window", m.get("ewin"))
            html += f'<td style="{cell_style}">{prop_html}</td>'

            if is_first:
                rc = ret_color(p["retention"])
                ret_cell_style = f"border-top:{top_border}; padding:10px 12px; vertical-align:middle; text-align:center; border-right:0.5px solid #e0e0e0; border-bottom:0.5px solid #e0e0e0;"

                # Connects configuration matrix and supplementary listings smoothly
                html += f'<td rowspan="{span}" style="{shared_style}">{config_html(p["config"], c_name, a_name, se_name, p.get("supplementary", {}))}</td>'
                testing_html = f'<div style="margin-bottom:2px;">{kv("Temp", p["temp"])}{kv("Pressure", p["pressure"])}<div style="border-top:0.5px dashed #e0e0e0; margin:6px 0;"></div>{kv("Mass Areal", p["areal_mg"])}{kv("Cap Areal", p["areal_mah"])}<div style="border-top:0.5px dashed #e0e0e0; margin:6px 0;"></div>{kv("C-Rate", p["crates"])}{kv("Voltage", p["voltage"])}</div>'
                html += f'<td rowspan="{span}" style="{shared_style}">{testing_html}</td>'
                retention_html = f'<div style="font-size:20px; font-weight:700; color:{rc}; text-align:center;">{p["retention"]}</div><div style="font-size:10px; color:#888; margin-top:3px; text-align:center;">@ {p["cycles"]} cycles</div>'
                html += f'<td rowspan="{span}" style="{ret_cell_style}">{retention_html}</td>'
                energy_html = f'{kv("Total Cycles", p["cycles"] + " cyc")}<div style="border-top:0.5px dashed #e0e0e0; margin:5px 0;"></div>{kv("Spec. Energy", p.get("energy"))}<div style="font-size:9px; color:#999; text-align:center; font-style:italic; margin-top:6px; border-top:0.5px dashed #e0e0e0; padding-top:4px;">Basis: Cell-level</div>'
                html += f'<td rowspan="{span}" style="{shared_style}">{energy_html}</td>'

            html += "</tr>"
    return html

TABLE_CSS = """
<style>
  .btbl * { box-sizing:border-box; margin:0; padding:0; font-family:-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif; }
  .btbl { overflow-x:auto; padding:1rem; background:#fff; }
  .btbl table { border-collapse:collapse; width:100%; min-width:1150px; border:0.5px solid #e0e0e0; }
  .btbl th { background:#f7f6f3; font-size:11px; font-weight:500; color:#555; padding:9px 12px; border:0.5px solid #e0e0e0; text-align:left; }
  .btbl .gh { text-align:center; font-size:12px; font-weight:600; color:#1a1a18; background:#f4f3ef; }
  .btbl td { font-size:11px; color:#1a1a18; }
</style>
"""

def build_table(results):
    return TABLE_CSS + f"""
<div class="btbl">
<table>
  <thead>
    <tr>
      <th rowspan="2" style="width:14%; vertical-align:bottom;">Paper</th>
      <th colspan="2" class="gh">Materials Info</th>
      <th colspan="2" class="gh">System Level Architecture</th>
      <th colspan="2" class="gh">Performance Indices</th>
    </tr>
    <tr>
      <th style="width:13%">Material</th>
      <th style="width:16%">Modulus &amp; Conductivity</th>
      <th style="width:23%">Battery Configuration</th>
      <th style="width:14%">Testing Conditions</th>
      <th style="width:10%">Retention</th>
      <th style="width:10%">Cycles &amp; Energy</th>
    </tr>
  </thead>
  <tbody>{build_rows(results)}</tbody>
</table>
</div>"""

# --- Pipeline Execution Loop ---
final_results = []
for f in my_files:
    try:
        print(f"Scanning {f} ...")
        final_results.append(battery_expert_scan(f))
        print("  ✓ Done")
    except Exception as e:
        print(f"  ✗ Error: {e}")

# Render out to the dynamic interface canvas
display(HTML(build_table(final_results)))

with open("battery_table.html", "w", encoding="utf-8") as f:
    f.write(build_table(final_results))
print("Saved → battery_table.html")
```
